```python
#importing all required  librraies
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import iplot
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#loading the sales dataset into pandas dataframe

in_file='Saless Data.csv'
sales_data=pd.read_csv(in_file)
sales_data
```

```python
#displaying the first few rows of thge dataframe to understand the structure 

print("Initial Data:")
print(sales_data.head())
```

```python
#information about the data

sales_data.info()
```

```python
sales_data.describe()
```

```python
#checking data type

sales_data.dtypes
```

```python
# Convert the 'Date' column to datetime for time-based analysis

sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
```

```python
# Group data by hour and calculate the total sales for each hour

hourly_sales = sales_data.groupby('Hour')['Sales'].sum().reset_index()
```

```python
# Visualize the best time for product advertisement using an interactive bar plot

# Set a beautiful style for the plot
sns.set(style="whitegrid", palette="pastel")

# Create a bar chart using seaborn
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='Hour', y='Sales', data=hourly_sales)
bar_plot.set(title='Best Time for Product Advertisement', xlabel='Hour of the Day', ylabel='Sales')

# Show the plot
plt.show()
```

```python
# Identify best-selling products

top_products = sales_data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Selling Products:")
print(top_products)
```

```python
# visualizing the top5 selling product

# Set the style for the plot
sns.set(style="darkgrid", palette="Set2")

# Create a bar chart using seaborn
plt.figure(figsize=(10, 5))
bar_plot = sns.barplot(x=top_products.index, y=top_products.values, palette='viridis')
bar_plot.set(title='TOP 5 SELLING PRODUCTS', xlabel='Product', ylabel='Total Quantity Ordered')

# Rotate x-axis labels for better readability
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='right')

# Show the plot
```

```python
sales_data.groupby('Product')['Price Each'].mean()
```

```python

products=sales_data.groupby('Product')['Quantity Ordered'].sum().index
quantity=sales_data.groupby('Product')['Quantity Ordered'].sum()
prices=sales_data.groupby('Product')['Price Each'].mean()
```

```python
# Set a Seaborn style
sns.set(style="whitegrid", palette="pastel")

# Create a figure with subplots
fig, ax1 = plt.subplots(figsize=(18, 10))

# Create the bar plot for quantity
sns.barplot(x=products, y=quantity, color='b', ax=ax1)

# Create the line plot for prices on the secondary y-axis
ax2 = ax1.twinx()
sns.lineplot(x=products, y=prices, color='r', ax=ax2)

# Rotate x-axis labels for better readability
ax1.set_xticklabels(products, rotation='vertical', size=12)

# Set labels and title
ax1.set_xlabel('Products', size=20)
ax1.set_ylabel('Quantity', color='b', size=20)
ax2.set_ylabel('Prices', color='r', size=20)
plt.title('PRODUCTS WITH THE HIGHEST REVENUE GENERATED AND VICE VERSA', fontweight='bold', fontsize=16)

# Show the plot
plt.show()
```

```python
# Group data by city and calculate the total purchase in each city

city_purchase = sales_data.groupby('City')['Sales'].sum().reset_index()
```

```python
# Find the city with the highest purchase

highest_purchase_city = city_purchase.loc[city_purchase['Sales'].idxmax()]
print(f"\nCity with Highest Purchase: {highest_purchase_city['City']}")
```

```python
# Set the style for the plot
sns.set(style="darkgrid")

# Create a bar chart using seaborn
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='City', y='Sales', data=city_purchase, palette='dark', ci=None)
bar_plot.set(title='TOTAL PURCHASE BY CITY', xlabel='City', ylabel='Sales')

# Rotate x-axis labels for better readability
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='right')

# Show the plot
plt.show()
```

```python
# Group data by product and calculate the total quantity sold for each product

product_quantity = sales_data.groupby('Product')['Quantity Ordered'].sum().reset_index()
```

```python
# Find the top 5 least purchased products

least_purchased_products = product_quantity.nsmallest(5, 'Quantity Ordered')

print("\nTop 5 Least Purchased Products:")
print(least_purchased_products)
```
